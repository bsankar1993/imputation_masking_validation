```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import json
import numpy as np

```

```python
with open('dmi_np_nrmse_a0.json') as f:
    dat = json.load(f)
df = pd.DataFrame.from_dict(dat, orient='index')
df['method']  = 'DMI'

with open('knn_30_nrmse_a0.json') as f:
    dat = json.load(f)
df2 = pd.DataFrame.from_dict(dat, orient='index')
df2['method']  = 'KNN'

with open('si_mean_nrmse_a0.json') as f:
    dat = json.load(f)

df3 = pd.DataFrame.from_dict(dat, orient='index')
df3['method']  = 'SI'
```

```python
all_df = pd.concat([df, df2, df3])
all_df['mask'] = all_df.index
all_df
```

```python
len(all_df['nrmse_individual'].iloc[11])
```

```python
# Create a new DataFrame for Seaborn plotting
plot_data = []
for idx, row in enumerate(all_df['nrmse_individual']):
    for val in row:
        plot_data.append({'nrmse_individual': val, 'mask': all_df['mask'][idx], 'method': all_df['method'][idx]})

df_plot = pd.DataFrame(plot_data)

# Plotting with Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_plot, x='mask', y='nrmse_individual', hue='method', ci='sd')

# Additional plot customization
plt.xlabel('mask')
plt.ylabel('nrmse_individual (mean)')
plt.title('Mean nrmse_individual with Error Bars')
plt.grid(True)
plt.legend(title='Method')

# plt.show()
plt.savefig('nrmse_line_error.pdf', dpi=300)
```

```python
# bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=df_plot, x='mask', y='nrmse_individual', hue='method')

# Additional plot customization
plt.xlabel('Mask')
plt.ylabel('nrmse_individual')
plt.title('nrmse_individual across Masks by Method')
plt.grid(True)
plt.legend(title='Method')

# plt.show()
plt.savefig('nrmse_bar_error.pdf', dpi=300)
```
